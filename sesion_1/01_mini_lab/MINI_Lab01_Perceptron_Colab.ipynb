## Contexto

El **perceptrón** es la unidad más básica de una red neuronal. Toma varias entradas, las pondera con pesos, las suma, y aplica una función de activación para tomar una decisión binaria.

### Escenario: ¿Voy al concierto esta noche?

Tienes tres factores que influyen en tu decisión:

| Variable | Significado |
|----------|-------------|
| `x1` | ¿Qué tanto dinero tengo? |
| `x2` | ¿Cuánto quieren ir mis amigos? |
| `x3` | ¿Qué tan probable es que llueva? |

> **Clave de este lab:** Las entradas y los pesos **no tienen que ser 0 o 1**.  
> Puedes usar cualquier número decimal para representar *qué tanto* se cumple cada condición.

---

## Parte 1 — Entradas (`x`)

Las entradas representan el **nivel** de cada factor, no solo si está presente o no.  
Usa valores decimales entre `0.0` y `1.0`:

- `0.0` → el factor no se cumple para nada  
- `0.5` → se cumple a medias  
- `1.0` → se cumple completamente

**Ejercicio 1:** Descomentar, asigna tus propios valores decimales y ejecutar.

```python
# x1 = 0.00   # ¿Qué tanto dinero tienes?          (0.0 = sin dinero, 1.0 = dinero suficiente)
# x2 = 0.00   # ¿Cuánto quieren ir tus amigos?     (0.0 = nadie quiere, 1.0 = todos entusiasmados)
# x3 = 0.00   # ¿Qué tan probable es que llueva?   (0.0 = cielo despejado, 1.0 = lluvia segura)

# print(f'Entradas: x1={x1}, x2={x2}, x3={x3}')
```

**✅ Solución Parte 1**

```python
x1 = 0.80   # Bastante dinero disponible
x2 = 0.30   # Pocos amigos quieren ir
x3 = 0.70   # Bastante probable que llueva

print(f'Entradas: x1={x1}, x2={x2}, x3={x3}')
```

---
## Parte 2 — Pesos (`w`) y Sesgo (`bias`)

Los pesos representan la **importancia** de cada factor en la decisión.  
Pueden ser positivos (favorecen la salida 1) o negativos (la penalizan).

El **sesgo** (`bias`) desplaza el umbral de activación. Un bias negativo hace al perceptrón más difícil de activar.

**Ejercicio 2:** Descomentar, asigna pesos decimales según la importancia de cada factor y ejecutar.

```python
# w1   = 0.00   # Peso del dinero        (positivo: el dinero ayuda a ir)
# w2   = 0.00   # Peso de los amigos     (positivo: los amigos motivan)
# w3   = 0.00   # Peso de la lluvia      (negativo: la lluvia desanima)
# bias = 0.00   # Sesgo                  (negativo: eres algo perezoso para salir)

# print(f'Pesos: w1={w1}, w2={w2}, w3={w3}  |  bias={bias}')
```

**✅ Solución Parte 2**

```python
w1   =  0.60   # El dinero es muy importante
w2   =  0.20   # Los amigos importan, pero no deciden solos
w3   = -0.40   # La lluvia es un factor negativo
bias = -0.30   # Sesgo negativo: eres algo perezoso para salir

print(f'Pesos: w1={w1}, w2={w2}, w3={w3}  |  bias={bias}')
```

---
## Parte 3 — Suma Ponderada (`z`)

La suma ponderada multiplica cada entrada por su peso y suma el sesgo:

$$z = (x_1 \cdot w_1) + (x_2 \cdot w_2) + (x_3 \cdot w_3) + bias$$

**Ejercicio 3:** Descomentar, implementar la fórmula y ejecutar.

```python
# z = ___  # completa la fórmula con x1, w1, x2, w2, x3, w3, bias

# print(f'Suma ponderada z = {z:.4f}')
```

**✅ Solución Parte 3**

```python
z = (x1 * w1) + (x2 * w2) + (x3 * w3) + bias
# z = (0.80*0.60) + (0.30*0.20) + (0.70*-0.40) + (-0.30)
# z = 0.48 + 0.06 - 0.28 - 0.30 = -0.04

print(f'Suma ponderada z = {z:.4f}')
```

---
## Parte 4 — Función de Activación

El perceptrón usa una **función de paso** para convertir `z` en una decisión binaria:

$$salida = \begin{cases} 1 & \text{si } z > 0 \\ 0 & \text{si } z \leq 0 \end{cases}$$

**Ejercicio 4:** Descomentar, implementar la función de paso y ejecutar.

```python
# if ___:        # condición sobre z
#     salida = ___
# else:
#     salida = ___

# print(f'Salida del perceptrón = {salida}')
# if salida == 1:
#     print('Decisión: Vas al concierto 🎵')
# else:
#     print('Decisión: Te quedas en casa 🏠')
```

**✅ Solución Parte 4**

```python
if z > 0:
    salida = 1
else:
    salida = 0

print(f'Salida del perceptrón = {salida}')
if salida == 1:
    print('Decisión: Vas al concierto 🎵')
else:
    print('Decisión: Te quedas en casa 🏠')
```

**Ejercicio 5:** Implementa los escenarios anteriores reutilizando tus fórmulas de las partes 3 y 4.

```python
# Escenario perfecto para salir
# x1, x2, x3 = 0.9, 0.8, 0.1
# z = ___
# salida = 1 if z > 0 else 0
# print(f'Escenario 1 → z={z:.4f}, salida={salida}')

# Muy mal tiempo
# x1, x2, x3 = 0.8, 0.7, 0.95
# z = ___
# salida = 1 if z > 0 else 0
# print(f'Escenario 2 → z={z:.4f}, salida={salida}')

# Solo y con poco dinero
# x1, x2, x3 = 0.2, 0.0, 0.3
# z = ___
# salida = 1 if z > 0 else 0
# print(f'Escenario 3 → z={z:.4f}, salida={salida}')
```

**✅ Solución Parte 5**

```python
# Pesos fijos (los mismos de Parte 2)
w1, w2, w3, bias = 0.60, 0.20, -0.40, -0.30

# Escenario 1: perfecto para salir
x1, x2, x3 = 0.9, 0.8, 0.1
z = (x1 * w1) + (x2 * w2) + (x3 * w3) + bias
# z = 0.54 + 0.16 - 0.04 - 0.30 = 0.36
salida = 1 if z > 0 else 0
print(f'Escenario 1 → z={z:.4f}, salida={salida}  → {"Vas al concierto 🎵" if salida else "Te quedas en casa 🏠"}')

# Escenario 2: muy mal tiempo
x1, x2, x3 = 0.8, 0.7, 0.95
z = (x1 * w1) + (x2 * w2) + (x3 * w3) + bias
# z = 0.48 + 0.14 - 0.38 - 0.30 = -0.06
salida = 1 if z > 0 else 0
print(f'Escenario 2 → z={z:.4f}, salida={salida}  → {"Vas al concierto 🎵" if salida else "Te quedas en casa 🏠"}')

# Escenario 3: solo y con poco dinero
x1, x2, x3 = 0.2, 0.0, 0.3
z = (x1 * w1) + (x2 * w2) + (x3 * w3) + bias
# z = 0.12 + 0.00 - 0.12 - 0.30 = -0.30
salida = 1 if z > 0 else 0
print(f'Escenario 3 → z={z:.4f}, salida={salida}  → {"Vas al concierto 🎵" if salida else "Te quedas en casa 🏠"}')
```

## Resumen

| Componente | Símbolo | Rol |
