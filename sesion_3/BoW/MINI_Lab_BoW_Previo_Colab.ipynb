### ⚠️ Lo que BoW NO puede hacer — y por qué importa

> **BoW no aprende, no generaliza y no predice.**

Mira esta pregunta: *¿La reseña "el cine no es bueno" es positiva o negativa?*

BoW te da vectores de conteos. Pero para **responder esa pregunta** necesitas algo más:

```
Texto → [BoW] → Vector numérico → [¿?] → Predicción
```

El `[¿?]` es un **modelo** — un algoritmo que aprende la relación entre  
los vectores y las etiquetas (positivo/negativo, spam/no spam, etc.)  
a partir de ejemplos conocidos.

Sin el modelo, BoW solo describe. **No decide nada por sí solo.**

---

### 🔗 La cadena completa de NLP

```
Texto crudo
    ↓
Preprocesamiento (limpiar, tokenizar)
    ↓
Representación  ← estás aquí (BoW)
    ↓
Mejora de representación  (TF-IDF → Lab 04)
    ↓
Representación semántica  (Word2Vec → Lab 05)
    ↓
Modelo (clasificador, regresor, red neuronal…)
    ↓
Predicción
```

> BoW es el **primer escalón** — necesario, pero nunca suficiente.
