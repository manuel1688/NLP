## 2. Definición del corpus

Definimos un texto de ejemplo en español para tokenizar.

```python
import spacy

# Cargar el modelo de español
# es_core_news_sm incluye: tokenizador, etiquetas POS, parser, NER
nlp = spacy.load("es_core_news_sm")

corpus = "Hola, mi nombre es Pinki Dagar. Soy estudiante de ingeniería en sistemas."

# Procesar el texto — genera un objeto Doc con todos los atributos
doc = nlp(corpus)

print("=== Corpus ===")
print(corpus)
```

## 3. Tokens

spaCy divide el texto en tokens automáticamente al procesarlo con `nlp()`.

```python
print("=== Tokens ===")
for token in doc:
    print(token.text)
```

## 4. Tokenización de oraciones

spaCy detecta los límites de oraciones automáticamente usando el parser de dependencias.

```python
print("=== Tokenización de oraciones ===")
for oracion in doc.sents:
    print(oracion.text)
```

## 5. Tokenización de palabras + atributos

Cada token en spaCy expone atributos adicionales sin necesidad de pasos extra:

| Atributo | Significado |
|---|---|
| `token.text` | Texto original del token |
| `token.pos_` | Categoría gramatical (sustantivo, verbo, etc.) |
| `token.is_punct` | `True` si es signo de puntuación |
| `token.is_stop` | `True` si es una stopword (palabra vacía) |

```python
print("=== Tokenización de palabras + atributos ===")
print(f"{'Token':<20} {'Categoría POS':<15} {'Puntuación':<12} {'Stopword'}")
print("-" * 58)
for token in doc:
    print(f"{token.text:<20} {token.pos_:<15} {str(token.is_punct):<12} {token.is_stop}")
```

## 6. Comparación: spaCy vs NLTK

| Característica | NLTK | spaCy |
|---|---|---|
| Facilidad para aprender | ✅ Ideal para comenzar | Requiere más contexto |
| Velocidad | Lento en textos grandes | ✅ Muy rápido |
| Atributos por token | Solo texto | ✅ POS, lema, NER, stopword, etc. |
| Uso en producción | Poco común | ✅ Muy usado en industria |
| Soporte multilenguaje | Limitado | ✅ Modelos para muchos idiomas |

```python
print("=== Comparación spaCy vs NLTK ===")
print("NLTK word_tokenize: separa puntuacion pero no da contexto semantico.")
print("spaCy: cada token trae categoria gramatical (POS), lema, si es stopword, si es puntuacion, etc.")
print("spaCy se usa mas en produccion; NLTK es mas adecuado para aprender los fundamentos.")
```
