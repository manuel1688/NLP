# Tokenization en NLP con NLTK

Este notebook muestra cómo realizar tokenización de oraciones y palabras usando NLTK en Python, incluyendo el uso de TreebankWordTokenizer.

## 1. Instalación de NLTK

Ejecuta la siguiente celda para instalar NLTK si no está disponible y descargar los recursos necesarios.

```python
# Instala NLTK y descarga recursos
!pip install -U nltk
import nltk
nltk.download('punkt')
```

## 2. Definición del corpus

Definimos un pequeño texto de ejemplo para tokenizar.

```python
CORPUS = """Hello, My name is Pinki Dagar.\nI am a student of b.tech computer science."""
print("=== Corpus ===")
print(CORPUS)
```

## 3. Tokenización de oraciones

Dividimos el texto en oraciones individuales usando `sent_tokenize`.

```python
from nltk.tokenize import sent_tokenize

documents = sent_tokenize(CORPUS)
print("=== Sentence Tokenization ===")
for sentence in documents:
    print(sentence)
```

## 4. Tokenización de palabras

Dividimos cada oración en palabras y signos de puntuación usando `word_tokenize`.

```python
from nltk.tokenize import word_tokenize

print("=== Word Tokenization ===")
for sentence in documents:
    print(word_tokenize(sentence))
```

## 5. Treebank Word Tokenizer

El `TreebankWordTokenizer` usa reglas del Penn Treebank y maneja la puntuación de forma especial.

```python
from nltk.tokenize import TreebankWordTokenizer

tokenizer = TreebankWordTokenizer()
print("=== Treebank Word Tokenizer ===")
print(tokenizer.tokenize(CORPUS))
```
