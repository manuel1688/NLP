# Tokenization in Natural Language Processing

Tokenization is the process of breaking down a text into smaller units called **tokens**. These tokens can be sentences, words, or even sub-words. It is a fundamental step in any NLP pipeline.

In this notebook, we will explore:
1. **Sentence Tokenization**: Breaking a corpus into individual sentences.
2. **Word Tokenization**: Breaking sentences into individual words.

## 1. Setup and Installation
First, we need to install the `nltk` (Natural Language Toolkit) library if it's not already available.

```python
pip install nltk
```

## 2. Defining the Corpus
A **corpus** is simply a large collection of text. Here, we define a simple multi-line string to work with.

```python
corpus = """Hello, My name is Pinki Dagar. 
I am a student of b.tech computer science."""
```

```python
print(corpus)
```

### Downloading NLTK Resources
We need the `punkt` and `punkt_tab` resources for sentence and word splitting.

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

## 3. Sentence Tokenization
This process converts a large paragraph or corpus into a list of individual sentences.

```python
from nltk.tokenize import sent_tokenize
documents = sent_tokenize(corpus)
```

Let's iterate through the tokenized sentences:

```python
for sentence in documents:
    print(sentence)
```

## 4. Word Tokenization
Word tokenization breaks sentences down into individual words and punctuation marks.

```python
from nltk.tokenize import word_tokenize
```

Applying word tokenization to each sentence in our documents:

```python
for sentence in documents:
    print(word_tokenize(sentence))
```

## 5. Treebank Word Tokenizer
The `TreebankWordTokenizer` uses conventions from the Penn Treebank, which handles punctuation (like periods at the end of sentences) differently than the standard `word_tokenize`.

```python
from nltk.tokenize import TreebankWordTokenizer
tokenizer = TreebankWordTokenizer()
tokenizer.tokenize(corpus)
```
