## 2. Ejemplo de lematización con spaCy

spaCy es la opción más moderna y práctica para lematización en español. Es rápida, lista para producción y fácil de usar.

```python
import spacy

nlp = spacy.load("es_core_news_sm")
doc = nlp("Los niños estaban corriendo rápidamente")

for token in doc:
    print(f"{token.text} -> {token.lemma_}")
```

**Ventajas de spaCy:**
- Rápida y lista para producción
- Soporta español muy bien
- No necesitas configurar mucho

## 3. Ejemplo de lematización con NLTK

NLTK es muy usada en enseñanza, pero menos en producción moderna. Su lematizador funciona mejor en inglés.

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
print(lemmatizer.lemmatize("running", pos="v"))  # Ejemplo en inglés
```

**Pros de NLTK:**
- Fácil para aprender

**Contras:**
- Funciona mejor en inglés que en español
- Necesita más configuración (etiquetado gramatical)

## 4. Ejemplo de lematización con Stanza

Stanza, desarrollada por Stanford, es muy precisa lingüísticamente, aunque más lenta que spaCy.

```python
import stanza

nlp_stanza = stanza.Pipeline('es')
doc = nlp_stanza("Los niños estaban corriendo")
for sentence in doc.sentences:
    for word in sentence.words:
        print(f"{word.text} -> {word.lemma}")
```

**Pros de Stanza:**
- Muy precisa lingüísticamente

**Contras:**
- Más lenta que spaCy

## 5. Comparación práctica entre las librerías

Veamos cómo lematizan la misma frase las tres librerías. Observa las diferencias y similitudes.

```python
frase = "Los niños estaban corriendo rápidamente"

# spaCy
print("spaCy:")
doc_spacy = nlp(frase)
for token in doc_spacy:
    print(f"{token.text} -> {token.lemma_}")

# NLTK (solo palabras en inglés)
print("\nNLTK:")
for word in ["running", "children"]:
    print(f"{word} -> {lemmatizer.lemmatize(word, pos='v')}")

# Stanza
print("\nStanza:")
doc_stanza = nlp_stanza(frase)
for sentence in doc_stanza.sentences:
    for word in sentence.words:
        print(f"{word.text} -> {word.lemma}")
```

**Resumen de la comparación:**
- **spaCy**: Mejor soporte para español, resultados listos para producción.
- **NLTK**: Útil para inglés, no recomendado para español.
- **Stanza**: Muy preciso, pero más lento.

## 6. ¿Por qué la lematización y el stemming perdieron relevancia?

La caída de la lematización y el stemming no fue por una sola razón, sino por la combinación de avances en NLP:

| **Factor**                | **Qué hace**                                      | **Por qué los reemplaza**                       |
|---------------------------|---------------------------------------------------|-------------------------------------------------|
| Tokenización subword      | Divide palabras (corriendo → corr + iendo)        | Captura raíces sin reglas manuales              |
| Embeddings contextuales   | Representan palabras según su contexto            | Capturan significado sin simplificar            |
| Transformers              | Aprenden relaciones del lenguaje                  | Integran semántica + gramática                  |

### 🔍 Ajuste clave sobre embeddings (lo importante)
- ❌ **No:** “se agrupan tridimensionalmente”
- ✔ **Sí:** Viven en espacios de alta dimensión y, durante el entrenamiento, palabras con usos similares terminan cerca entre sí.
- 👉 **Importante:**
    - El agrupamiento es consecuencia del entrenamiento, **no la causa** del entendimiento.

| Lematización    | ⚖️ Opcional     |
| Stemming        | ❄️ En declive   |
