## 3. Tokenización de oraciones

Dividimos el texto en oraciones individuales usando `sent_tokenize`.

````python
from nltk.tokenize import sent_tokenize

oraciones = sent_tokenize(corpus, language='spanish')
print("=== Tokenización de oraciones ===")
for oracion in oraciones:
    print(oracion)
````

## 4. Tokenización de palabras

Dividimos cada oración en palabras y signos de puntuación usando `word_tokenize`.

````python
from nltk.tokenize import word_tokenize

print("=== Tokenización de palabras ===")
for oracion in oraciones:
    tokens = word_tokenize(oracion, language='spanish')
    print(tokens)
````

## 5. Tokenizador Treebank

El `TreebankWordTokenizer` usa reglas del Penn Treebank y maneja la puntuación de forma especial. Se aplica al corpus completo directamente.

````python
from nltk.tokenize import TreebankWordTokenizer

tokenizador = TreebankWordTokenizer()
print("=== Tokenizador Treebank ===")
print(tokenizador.tokenize(corpus))
````
