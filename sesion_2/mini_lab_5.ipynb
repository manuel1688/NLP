## 3. Tokenizar y construir vocabulario único

```python
from nltk.tokenize import word_tokenize

all_tokens = word_tokenize(corpus, language='spanish')
vocab = sorted(set(t.lower() for t in all_tokens))

print('=== Tokens únicos ===')
print(vocab)
print(f'Total tokens: {len(all_tokens)}  |  Vocabulario: {len(vocab)}')
```

## 4. Asignar un vector aleatorio a cada token (dim=50)

```python
import numpy as np

EMBEDDING_DIM = 50
np.random.seed(42)

# Diccionario: token -> vector aleatorio
embeddings = {token: np.random.randn(EMBEDDING_DIM) for token in vocab}

print(f'=== Embedding aleatorio (dim={EMBEDDING_DIM}, primeros 5 valores) ===')
for token, vec in embeddings.items():
    vals = ', '.join(f'{v:.2f}' for v in vec[:5])
    print(f'  {token:>15s} -> [{vals}, ...]')
```

## 5. Reducir de 50D a 3D con PCA

```python
from sklearn.decomposition import PCA

matrix = np.array([embeddings[t] for t in vocab])
pca = PCA(n_components=3)
coords_3d = pca.fit_transform(matrix)
```

## 6. Graficar en 3D

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(
    coords_3d[:, 0],
    coords_3d[:, 1],
    coords_3d[:, 2],
    c='steelblue',
    s=80,
    edgecolors='black',
    linewidths=0.5,
)

for i, token in enumerate(vocab):
    ax.text(
        coords_3d[i, 0],
        coords_3d[i, 1],
        coords_3d[i, 2],
        f'  {token}',
        fontsize=9,
    )

ax.set_title('Embedding aleatorio (sin entrenamiento) — PCA 3D', fontsize=14)
ax.set_xlabel('Componente 1')
ax.set_ylabel('Componente 2')
ax.set_zlabel('Componente 3')

output_path = 'embedding_random.png'
fig.savefig(output_path, dpi=150, bbox_inches='tight')
print(f'=== Gráfica guardada en: {output_path} ===')
plt.show()
```

## 7. Explicación

Los vectores fueron asignados al azar (`np.random.randn`), por eso los tokens aparecen dispersos sin ninguna agrupación semántica:

- Sustantivos no están cerca aunque sean de la misma categoría.
- `,` y `.` no se agrupan aunque ambos sean puntuación.
